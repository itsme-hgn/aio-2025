## Concat two dataframes

```python
import pandas as pd

df1 = pd.read_csv('advertising_p1.csv')
df1
```

```python
df2 = pd.read_csv('advertising_p2.csv')
df2
```

```python
df3 = pd.concat([df1, df2])
df3
```

```python
df3 = df3.reset_index(drop=True)
df3
```

```python
df4 = pd.concat([df1, df2], ignore_index=True)
df4
```
